The National Map has an API for data access and download. The National Elevation Dataset 1/9 arc-second products were selected. The results were filters with a bounding box using these parameters, `-81.1139,35.3845,-78.7408,36.6111`. 

```python
# url for API
url = 'https://viewer.nationalmap.gov/tnmaccess/api/products?datasets=National+Elevation+Dataset+%28NED%29+1%2F9+arc-second&bbox=-81.1139%2C35.3845%2C-78.7408%2C36.6111&q=&dateType=dateCreated&start=&end=&polyCode=&polyType=&offset=&max=&outputFormat=JSON&_csrf=1155e4b6-3d4d-409c-8058-99083bab2c28'

# request data
response = requests.get(url)
```

```python
# preview data
response.json()
```

```python
print('The API from the National Map returned {} features'.format(response.json()['total']))
```

These features represent Digitial Elevation Models or DEMs for the CommunityViz study area. The DEMs will be used to identify areas where the slope is greater than 20%. Those steep slope areas will be vectorized and added to the *constrained_areas* layer in the CommunityViz model.
